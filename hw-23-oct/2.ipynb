```python
import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 1
t_end = 10
```

```python
def solve_Euler(dt):
    global g, t_end
    x = [0]
    vx = [0]
    for t_now in np.arange(dt, t_end + dt, dt):
        x.append(x[-1] + vx[-1] * dt)
        vx.append(vx[-1] + g * dt)
    return x[-1]
```

```python
def solve_Euler_Cauchy(dt):
    global g, t_end
    x = [0]
    vx = [0]
    for t_now in np.arange(dt, t_end + dt, dt):
        vx.append(vx[-1] + g * dt)
        x.append(x[-1] + (vx[-2] + vx[-1]) / 2 * dt)
    return x[-1]
```

```python
def solve_an():
    global g, t_end
    return g * (t_end ** 2) / 2
```

```python
e = []
ec = []
dts = []
for k in range(10 + 1):
    dt = 0.1 / (2 ** k)
    e.append(np.abs(solve_an() - solve_Euler(dt)))
    ec.append(np.abs(solve_an() - solve_Euler_Cauchy(dt)))
    dts.append(dt)
```

```python
plt.plot(dts, e, label='Euler method')
plt.plot(dts, ec, label='Euler-Cauchy method')
plt.title('ksi(dt)')
plt.xlabel('dt')
plt.ylabel('ksi')
plt.grid()
plt.legend()
```
