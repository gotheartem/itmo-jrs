```python
import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(x_init, y_init):
    x = [x_init]
    y = [y_init]
    t = [0]
    dt = 1e-4
    t_end = 100
    for t_now in tqdm.tqdm(np.arange(dt, t_end + dt, dt)):
        x.append(x[-1] + dt * (-y[-1] + x[-1] * (1 - (x[-1] ** 2 + y[-1] ** 2) ** 0.5)))
        y.append(y[-1] + dt * (x[-1] + y[-1] * (1 - (x[-1] ** 2 + y[-1] ** 2) ** 0.5)))
        t.append(t_now)

    plt.plot(x, y, linestyle='', marker='.', markersize=0.5)
    plt.title('y(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()

    plt.plot(t[-100000:], x[-100000:])
    plt.title('x(t)')
    plt.xlabel('t')
    plt.ylabel('x')
    plt.grid()
    plt.show()

    plt.plot(t[-100000:], y[-100000:])
    plt.title('y(t)')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.grid()
    plt.show()
```

```python
# a)
plot(0, 0.4)
```

```python
# b)
plot(1, 0)
```

```python
# c)
plot(1.2, 0.6)
```
